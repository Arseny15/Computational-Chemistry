```python
# modified Shields and el. example code
from edbo.bro import BO_express
    
reaction_components={
    'reagent':['O=C([O-])C.[K+]','O=C([O-])C(C)(C)C.[K+]',
               'O=C([O-])C.[Cs+]', 'O=C([O-])C(C)(C)C.[Cs+]'],
    'solvent': ['CC(N(C)C)=O', 'CCCC#N' ,
                'CCCCOC(C)=O', 'CC1=CC=C(C)C=C1'],
    'catalyst_one':['P(C1=CC=CC=C1)(C2CCCCC2)C3=CC=CC=C3',
                    'C1(P(C2CCCCC2)C3CCCCC3)=CC=CC=C1C4=CC=CC=C4'],
    'solvent_conc': [263.1579, 150, 98.039215],
    'temperature': [120, 105, 90]}

encoding={'reagent':'smiles',
          'solvent':'smiles',
          'catalyst_one':'smiles',
          'solvent_conc':'numeric',
          'temperature':'numeric'}

bo = BO_express(reaction_components=reaction_components, 
                encoding=encoding,
                batch_size=7,
                init_method='PAM',
                acquisition_function='TS',
                target='yield')

bo.init_seq.visualize = False
bo.init_sample()                    
bo.get_experiments(structures=True)
```

```python
bo.export_proposed('init.csv') 
```

```python
bo.save()
```

```python
from edbo.bro import BO_express

bo = BO_express()
bo.load()
```

```python
bo.get_experiments(structures=True)
```

```python
bo.add_results('results/init.csv')
```

```python
bo.run()
```

```python
bo.model.regression()
```

```python
bo.export_proposed('round0.csv')
```

```python
#modified from Shields and al. example code
def workflow(export_path):
    bo.plot_convergence()
    bo.model.regression()
    bo.export_proposed(export_path)
```

```python
bo.add_results('results/round0.csv')  
bo.run()
workflow('round1.csv')
```
